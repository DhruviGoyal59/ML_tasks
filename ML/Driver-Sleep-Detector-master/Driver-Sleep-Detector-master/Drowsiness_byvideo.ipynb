```python
import cv2

# Load Haar Cascade classifiers for face and eye detection
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Open the video file for reading (change 'video.mp4' to your video file)
video_file = 'Downloads\\driver3.mp4'
cap = cv2.VideoCapture(video_file)

# Initialize variables
drowsy_frames = 0
CONSECUTIVE_FRAMES_THRESHOLD = 15  # Adjust this value as needed

while True:
    # Capture a frame from the video file
    ret, frame = cap.read()

    # Exit the loop if the video is finished
    if not ret:
        break

    # Convert the frame to grayscale for face and eye detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

    for (x, y, w, h) in faces:
        # Draw a blue rectangle around the detected face
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)

        # Region of interest for eyes within the face region
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        # Detect eyes within the face region
        eyes = eye_cascade.detectMultiScale(roi_gray)

        if len(eyes) == 0:
            drowsy_frames += 1
            if drowsy_frames >= CONSECUTIVE_FRAMES_THRESHOLD:
                # Perform an action when drowsiness is detected (e.g., print a message)
                print("Driver is drowsy!")
        else:
            drowsy_frames = 0
            print("Driver is not drowsy!")

        for (ex, ey, ew, eh) in eyes:
            # Draw green rectangles around detected eyes
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

    # Display the frame with rectangles
    cv2.imshow('Drowsiness Detection', frame)

    # Exit the loop when 'q' is pressed or when the video ends
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close OpenCV windows
cap.release()
cv2.destroyAllWindows()
```
